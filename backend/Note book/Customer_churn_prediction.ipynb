This Python code is designed to predict whether a new customer will churn using a pre-trained machine learning model. First, the trained model and the training column information are loaded from files using joblib. The training columns include both numeric features such as Age, Tenure, MonthlyCharges, and TotalCharges, as well as one-hot encoded categorical features like Gender, Contract type, and PaymentMethod. Next, a new customer’s information is provided as a dictionary, where each value is stored in a list to make it compatible with Pandas DataFrame. A DataFrame is then created for this new customer, initially including only the numeric features. To handle the categorical variables, all one-hot encoded columns from the training data are added to the DataFrame with a default value of zero. The specific values for the new customer’s Gender, Contract, and PaymentMethod are then updated to one in the corresponding columns to correctly reflect their category. After that, the DataFrame is aligned with the training columns, ensuring it has the exact same structure that the model expects. Once the DataFrame is prepared, the model is used to make a prediction: a value of 0 indicates the customer is not likely to churn, while a value of 1 indicates the customer is likely to churn. Additionally, if the model supports probability predictions, the code calculates the probability of churn, which provides insight into the confidence of the prediction. Finally, the prediction and, optionally, the churn probability are printed for interpretation. Overall, this code automates the process of preparing new customer data, encoding categorical features, aligning with the model’s expected input, and predicting churn in a structured and reproducible way.
